```python
# Input: Total minutes
total_minutes = int(input("Enter Total Minutes:"))
#Calculate Days
days = total_minutes // 1440           # 1 day = 1440 minutes
# Calculate remaining Minutes After Days
remaining_minutes = total_minutes % 1440
#Calculate Hours
hours= remaining_minutes // 60   # 1 hour = 60 minutes
#Calculate Remaining Minutes After Hours
minutes = remaining_minutes % 60
print(f"{days} day(s), {hours} hours(s), and {minutes} minutes(s)")
```
